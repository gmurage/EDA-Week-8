# Non Graphical Exploratory Data Analyis (NGEDA)
NGEDA provides an idea about the description and distribution of the variables. Univariate NGEDA is a principal form of data analysis that involves only one variable to identify underlying data distribution and the characteristics of population distribution. It also covers outlier detection and for any quantitative variable univariate EDA helps in making initial assessments on the variable distribution using the data sample. NGEDA are objective and quantitative in nature failing to give complete representation of data (Banu et al., 2022).Univariate means only one variable, exposure or outcome is under study (Data et al., 2016). There are two types of NGEDA: univariate NGEDA and multivariate NGEDA.

# Univariate Analysis
<BR>
The prefix "Uni" means one. It means that univariate analysis is the analysis of one variable at a time. For numeric features we want to know the range of values present and how often these values of groups of values do indeed occur. On the other hand for categorical features, we seek to know the number of unique classes and how frequently they occur. Numeric features are variables with numbers  and mathematical operations can be performed on them. Numerical variables can either be discreet or continuous or ratios. Discreet means that they are countable integers with clear boundaries. Continuous means that they can take any value even decimals within a range. According to Mukhiya and Ahmed (2020), the simplest form of analyzing data is called univariate analysis, it does not deal with causal relationships but rather  deals with taking data, summarizing it and finding patterns among the values. We first summarize the data for each numeric feature using df.describe, This gives us the actual statistics and we can even take it a step further and the the 95th and 5th percentile to get a sense of the outliers in the data. While summarization is very helpful, it is not enough and we must get a better intuitive understanding of the distribution by graphically visualizing using various tools.

# Types of Univariate NGEDA
1. Central Tendency: In a population distribution the central tendency measures mean, median and mode whereby the median is preferred for skewed distribution or when there are outliers.

2. Spread: Spread indicates  how far from the center one can find the data value. Variance, interquartile range and standard deviation are the common measures used to find spread. In a distribution. The variance is computed  by taking the mean of the squares of all the individual deviations. Standard deviation on the other hand is the square root of the variance.

3. Skewness is the measure of asymmetry of the variable in the data set about the means. Skewness can either be positive or negative. If mean is higher than the median it is a positive skew which has a longer tail on the right-hand side. If mean and median are the same it is a symmetrical distribution with both left and right tails being the same, and if the mean is less than the median it is a negative skew with the left-hand tail being longer than the right (Mukhiya & Ahmed, 2020)

4. Kurtosis is a measure of the heaviness of the distribution tails. Kurtosis is a statistical measure that illustrates  how heavily the tails of the distribution differ from those of a normal distribution. It helps identify whether a given distribution contains extreme values. It is a measure of outlier presence in a given distribution. Both high and low kurtosis is an indicator that data needs further investigation. The higher the kurtosis the higher the outliers. Mesokurtic is when kurtosis is equal to zero and there is a normal distribution, leptokurtic has a kurtosis greater  than 3 indicating that the distribution produces more outliers. Platykurtic has negative kurtosis and very thin compared to the normal distribution (Mukhiya & Ahmed, 2020)

# Finding outliers and dealing with them
We shall use the interquartile range method to calculate data points that fall out side of 1.5 times of the inter quartile range (IQR) (Q3-Q1). Therefore + and - 1.5*IQR means that we are only  considering data within the constraints. We ran the describe command afterwards to make sure that the changes took place and data transformation took place.
<BR>
    <BR>
        <BR>

```python
# Checking for outliers using the interquartile range method(25th and 75th percentiles) and removing the whole record if an outlier appears
# Import the library
import pandas as pd

# Assuming df is a DataFrame with numerical columns
df = df.astype(float)  # Convert all columns to float type

Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1

df = df[~((df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))).any(axis=1)]
```

```python
# validatign that the changes have taken place after dropping the outliers
df.describe ()
```

# Histogram
Histograms are some of the most useful EDA techniques, and allow you to gain insight into your data, including distribution, central tendency, spread, modality and outliers. Histograms are bar plots of counts versus subgroups of an exposure variable. Each bar represents the frequency (count) or proportion (count divided by total count) of cases for a range of values. The range of data for each bar is called a bin. Histograms give an immediate impression of the shape of the distribution. You can tell if the data is  symmetrical, unimodal, multimodal, skewed, or has outliers. The number of bins heavily influences the final aspect of the histogram; a good practice is to specify bin values  generally from 10 to 50 (Data et al., 2016).

## Identifying the Optimum Number of Clusters- KMeans
There are a number of ways to identify the optimum number of clusters. The first most popular method to find the optimum value of k is is known as the Elbow method. The Elbow method allows us to plot the inertia, which is a measure of how well the data was clustered by the K-Means algorithm, against the number of clusters. From the plot, we are looking for a point where the inertia begins to slow.

We deem this method as not the most effective method to do do. In our algorithm, instead, we initialize random centroids to iterate through each column in the data and then randomly sample a single value. Each centroid will be similar to a row with values but they need to be random from our data set.The apply method in the data frame will iterate through each column in the data then randomly select a single X variable and turn it into a float. One could ideally assign centroids 1 to 10 values versus using this apply method that we chose "centroid= data.apply(lambda x: float(x.sample()))".

Kmeans algorithim is an unsupervised machine learning technique that allows us to cluster data points, allowing us to find patterns that help us analyze it more effectively. K means is an algorithm of inputting data and letting the machine look for trends versus predicting the the trends. It is an iterative process which means that it will converge to the optimal clustering over time.

K means divides the data into various clusters and the number of clusters is equal to the value of K. That is if the value of k is = 3, then the data will be divided into 3 clusters, each value of K is a centroid around which the data points will gather.

Distance calculation can be done by any method but we will use the Euclidean method for distance measurement, which is the distance between two points (x1, y1) and (x2, y2). Square x2-x1 add to square of y2-y1 and find the square root of the sum.

Pros of K means Clustering:
1. A relatively simple algorithm to apply
2. Flexible and works well with large data
3. Convergence is guaranteed

Cons of K means Clustering:
1. We have to manually define the number of centroids
2. It is not immune to outliers
3. It depends on the initial values of centroid chosen 

To run a k-means clustering:
1. Specify the number of clusters one wants(usually referred to as k)
2. Randomly initialize the centroid for ech cluster. The centroid is the datapoint that is in the center of the cluster.
3, Determine which data points belong to which cluster by finding the closest centroid to each data point.
4. Update the centroids based on the geometric mean of all the data points in the cluster.
5. Run 3 and 4 until the centroids stop changing. Each run is referred to as an iteration.

At iteration 2 the K means clsustering has divided the census 202o data into 6 clusters based on the similarity of each data points within each cluster. The algorithm began by randomly selecting a number of data points as the initial cluster centroids. It then assigned each data point to the closest cluster centroid based on the euclidean distance between the data point and the centroid . Next the K algorithim recalculated the centroids for each cluster by taking the mean of all the data points in the cluster. It reiterated the process of reassigning data points to the closest centroid and recalculating the centroids  was then repeated until the clusters stabilized and the centroids stopped shifting.
